```python
import pandas as pd
import numpy as np

#更改部分字段名字，使其为英文且符合学术规范
df = pd.read_csv(r'C:\Users\ASUS\Desktop\Master Thesis\Data of My Thesis\Education Data\important!!\Process Check\all_data_final_with_IMR.csv')
print("数据加载成功。样本量:", len(df))
print("\n" + "="*50 + "\n")

# 打印原始列名
print("原始字段名称：")
print(df.columns.tolist())
print("\n" + "="*50 + "\n")

# 定义重命名映射字典
rename_mapping = {
    'Lib_DREN': 'Department',
    'Ecole': 'School',
    'Statut': 'SchType',
    'Milieu_ecole': 'SchLocation',
    'Clas': 'Grade',
    'sexe': 'Gender',
    'date': 'Date',
    'rain_hist_vol': 'RainHistVol',
    'Heavy_Rain': 'HeavyRain',
    'Low_Rain': 'LowRain',
    'Extreme_Fluct_Month': 'ExtreFluctMonth',
    'Extreme_Fluct_Intra': 'ExtreFluctIntra',
    'ADM1_FR': 'Region',
    'ADM1_PCODE': 'RegionCode',
    'School_Year': 'SchYear'
}

# 应用重命名
df.rename(columns=rename_mapping, inplace=True)

# 打印重命名后的列名
print("重命名后的字段名称：")
print(df.columns.tolist())
print("\n" + "="*50 + "\n")
```

# 加入一级行政区：District(大区)

```python
df_dist = pd.read_excel(r'C:\Users\ASUS\Desktop\Master Thesis\Data of My Thesis\Education Data\important!!\Source Data for Python Analyse\Region for District.xlsx')
df = df.merge(df_dist[['RegionCode', 'District']], on='RegionCode', how='left')

print("District 映射完成。前5行预览：")
print(df[['RegionCode', 'District']].head())
print("\n" + "="*50 + "\n")
```
